```python
from sklearn.model_selection import train_test_split,cross_val_score,KFold
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import time
```

```python
x=df1.iloc[:,1:]
y=df1.iloc[:,0]
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=100,test_size=0.70)
```

```python
lm=LinearRegression()
start_time = time.time()
model=lm.fit(x_train,y_train)
end_time = time.time()
egitim_sure=end_time-start_time
print(f"Eğitim süresi: {egitim_sure} saniye")
```

```python
model.score(x_test,y_test)
```

```python
start_ev_time=time.time()
tahminler=model.predict(x_test)
end_ev_time=time.time()
tahmin_sure=end_ev_time-start_ev_time
print(f"Tahmin süresi: {tahmin_sure} saniye")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
import numpy as np
# y_test: Gerçek değerler, y_pred: Tahmin edilen değerler
mae = mean_absolute_error(y_test,tahminler)
mse = mean_squared_error(y_test, tahminler)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((y_test - tahminler) / y_test)) * 100
r2_degeri = r2_score(tahminler, y_test)

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"MAPE: {mape}")
print("R² değeri:", r2_degeri)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
hidden_units1 = 128  # İlk gizli katmanda 256 nöron
hidden_units2 = 64  # İkinci gizli katmanda 64 nöron
learning_rate = 0.0001

model4 = Sequential()
model4.add(Dense(hidden_units1, kernel_initializer='normal', activation='relu',input_dim=x_train.shape[1]))
model4.add(Dropout(0.3))  # Aşırı uyumu önlemek için Dropout
model4.add(Dense(hidden_units2, kernel_initializer='normal', activation='relu'))
model4.add(Dropout(0.3))
model4.add(Dense(1, kernel_initializer='normal', activation='linear'))  # Regresyon için çıkış katmanı
model4.compile(optimizer="adam", loss="mse")
```

```python
# Kayıp Fonksiyonu ve Learning Rate tanımlanıyor.
optimizer = torch.optim.Adam(model10.parameters(), lr=1e-3)
loss_fn = nn.MSELoss()

#Epoch büyüklüğü ve ekrandaki değerlendirme metrikleri tanımlanıyor.
epochs = 50
epoch_losses = []
epoch_val_losses = []
epoch_r2_scores = []
# Test verilerinin kopyalanması (Değişimlerin önüne geçmek için)
x_categ_test = x_categ_test.clone()
x_cont_test = x_cont_test.clone()
y_test = y_test.clone()
model10.train()
start_time = time.time()

for epoch in range(epochs):
    epoch_loss = 0
    all_predictions, all_targets = [], []

    for x_categ_batch, x_cont_batch, y_batch in train_loader:
        optimizer.zero_grad()


        predictions = model10(x_categ_batch, x_cont_batch)


        loss = loss_fn(predictions, y_batch)
        loss.backward()
        optimizer.step()


        epoch_loss += loss.item()

        all_predictions.extend(predictions.detach().cpu().numpy())
        all_targets.extend(y_batch.cpu().numpy())

    epoch_r2 = r2_score(all_targets, all_predictions)
    epoch_losses.append(epoch_loss/len(train_loader))
    epoch_r2_scores.append(epoch_r2)

    # Doğrulama aşaması
    model10.eval()  # Değerlendirme moduna geçiş
    val_loss = 0
    val_targets, val_predictions = [], []

    with torch.no_grad():
        for x_categ_batch, x_cont_batch, y_batch in test_loader:
            # Test verileriyle tahmin yapılır
            predictions = model10(x_categ_batch, x_cont_batch)

            # Kayıp hesaplanır
            loss = loss_fn(predictions, y_batch)
            val_loss += loss.item()

            # Tahminler ve hedefler toplanır
            val_predictions.extend(predictions.detach().cpu().numpy())
            val_targets.extend(y_batch.cpu().numpy())

    val_loss /= len(test_loader)  # Ortalama doğrulama kaybı
    epoch_val_losses.append(val_loss)

    # Val R² Skoru (opsiyonel)
    val_r2 = r2_score(val_targets, val_predictions)

    print(f"Epoch {epoch + 1}/{epochs}, "
          f"Loss: {epoch_loss / len(train_loader):.4f}, "
          f"Val Loss: {val_loss:.4f}, "
          f"R² Score: {epoch_r2:.4f}, "
          f"Val R²: {val_r2:.4f}")


end_time = time.time()
print(f"Training completed in {end_time - start_time:.2f} seconds")
```
